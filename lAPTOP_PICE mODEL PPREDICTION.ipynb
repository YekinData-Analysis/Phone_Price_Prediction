```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#we traning 70% of the dataset 
```

```python
models={'DTRegressor':DecisionTreeRegressor(),
       'BRegressor':BaggingRegressor(),
       'KNNRegressor':KNeighborsRegressor()}

#instancetiating The Models
```

```python
Results={'model':[],
        'Train_Score':[],
        'Test_Score':[],
        'RMSE':[]}

#creating a function 
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

#importing the evaluation metrics
```

```python
#geting The lenght of the X_train

len(X_train)
```

```python
#here We Are, 

for name,model in models.items(): #we create 4 looops
    model.fit(X_train,y_train)  # we fit in
    train_score=model.score(X_train,y_train) #we get The Traing Score
    test_Score=model.score(X_test,y_test) #we getting the test scores
    test_predict=model.predict(X_test) #getting The prediction
    RMSE=mean_absolute_error(test_predict,y_test) #the evalutaion scores
    Results['model'].append(name) #  then we append We The function we created
    Results['Train_Score'].append(train_score)
    Results['Test_Score'].append(test_Score)
    Results['RMSE'].append(RMSE)
    print('MOdel:',name) #then We Print
    print('Train:',train_score)
    print('Test Score:',test_Score)
    print('RMSE:',round(RMSE,2))
    print('=========================')
```
